# Importing the libraries

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

# Binary Frequency Shift Keying(BFSK)

## Wave parameters

```python
fc1 = int(input("Enter frequency of carrier signal 1: "))
fc2 = int(input("Enter frequency of carrier signal 1: "))
fm = int(input("Enter frequency of modulating signal: "))
A = int(input("Enter amplitude of carrier signal: "))
```

## Sampling rate

```python
time = int(input("Enter number of samples per bit of data sequence"))
```

```python
t = np.linspace(start=0, stop=1, num=time, endpoint=True)
```

## Modulating signal

```python
b = np.array(signal.square(2*np.pi*fm*t)).T
```

```python
u = [int(x+1) if x == -1 else int(x) for x in b]
```

## Carrier signals

```python
carrier1 = np.array(A*np.sin(2*np.pi*fc1*t)).T
```

```python
carrier2 = np.array(A*np.sin(2*np.pi*fc2*t)).T
```

## Modulated BFSK signal

```python
m = []
for i in range(len(u)):
    if u[i] == 1:
        m.append(carrier1[i])
    if u[i] == 0:
        m.append(carrier2[i])
```

# Plotting the results

```python
fig, axes = plt.subplots(nrows=4, ncols=1)
axes[0].plot(t, u)
axes[1].plot(t, carrier1)
axes[2].plot(t, carrier2)
axes[3].plot(t, m)

axes[0].set_title('Unipolar modulating signal')
axes[1].set_title('Carrier signal 1')
axes[2].set_title('Carrier signal 2')
axes[3].set_title('BFSK signal')

plt.tight_layout()

fig.set_figheight(12)
fig.set_figwidth(20)

axes[0].set_xlabel('time')
axes[0].set_ylabel('Amplitude')
axes[1].set_xlabel('time')
axes[1].set_ylabel('Amplitude')
axes[2].set_xlabel('time')
axes[2].set_ylabel('Amplitude')
axes[3].set_xlabel('time')
axes[3].set_ylabel('Amplitude')

axes[0].grid(which='major', axis='both')
axes[1].grid(which='major', axis='both')
axes[2].grid(which='major', axis='both')
axes[3].grid(which='major', axis='both')
```
